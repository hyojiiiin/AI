```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split

df = pd.read_csv("C:/Users/hyoji/Downloads/titanic.csv")

# 데이터프레임 확인
df
```

```python
# 결측치 처리
df["Age"] = df["Age"].fillna(df["Age"].mean())  # 평균 나이로 대체
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])  # 최빈값으로 대체

# 불필요한 컬럼 제거
df.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"], inplace=True)

# 범주형 데이터 인코딩 (Sex, Embarked)
df = pd.get_dummies(df, columns=["Sex", "Embarked"], drop_first=True)

# 레이블 분포 확인 (imbalanced 여부)
label_counts_before = df["Survived"].value_counts()
df.head(), label_counts_before
```
